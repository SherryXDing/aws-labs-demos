```python
from tool import darknet2pytorch
import torch
```

## Load weights from darknet format
### For yolov3, Remember to add "scale_x_y = 1.00" to each [yolo] layer in the .cfg file

```python
model = darknet2pytorch.Darknet('../yolov3/cfg/yolov3.cfg', inference=True)
model.load_weights('../yolov3/yolov3.weights')
```

## Save weights to pytorch format

```python
torch.save(model.state_dict(), '../yolov3/yolov3.pth')
```

## Reload weights from pytorch format

```python
model_pt = darknet2pytorch.Darknet('../yolov3/cfg/yolov3.cfg', inference=True)
model_pt.load_state_dict(torch.load('../yolov3/yolov3.pth'))
```
